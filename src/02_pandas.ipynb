```python
import pandas as pd
```

```python
#df=data frame
df = pd.read_excel(r"D:\Python_Project\src\data\intro_pandas\employee.xlsx")
print(df)
```

```python
#df=data frame
df2 = pd.read_excel(r"D:\Python_Project\src\data\intro_pandas\employee.xlsx")
print(df2)
```

# Concat

#### การ join table

```python
df_total = pd.concat([df,df2])
df_total
#คอลัมน์ต้องชื่อเหมือนกันเป๊ะไม่งั้นจะรวมไม่ได้
```

#### Reset index

```python
#เอา data frame ที่รวมกันไป reset index
df_reset_index = df_total.reset_index(drop=True)
df_reset_index
```

# Drop Duplicate

```python
boolean = df_reset_index['emp_name'].duplicated().any()
print(boolean)
print(len(df_reset_index))
#ต้องการเอาข้อมูล Dup ออกจากผลลัพธิ์
df_dup = df_reset_index.drop_duplicates(["emp_name"])
print(len(df_dup))
boolean = df_dup['emp_name'].duplicated().any()
print(boolean)
df_dup
```

# Drop Null

```python
boolean = (df_dup.isnull().values.any())
print(len(df_dup))
#ต้องการเอาข้อมูล Null NaN ออกจากผลลัพธิ์
df_null = df_dup.dropna()
print(len(df_null))
df_null
```

# Calculate DataFrames

```python
import datetime

today = datetime.date.today()
year = datetime.date.today().year
print(today)
print(today.year)
print(today.month)
print(today.day)

df_age = df_null.assign(age= year-df_null[["birthday year"]])
print(df_age)
```

```python
df_age[["name","surname"]] = df_age["emp_name"].str.split(expand=True)
df_age
#defaul split คือช่องว่าง ถ้าต้องการอย่างอื่นให้ใส่เอง
```

# Custom Columns

```python
df_col = df_age[["name","surname","age","country"]]
#การเลือกเอาคอลัมน์ที่ต้องการมา SHOW
df_col
```

# Edit Columns name

```python
df_edit_name = df_col.rename(columns={'name' : "first_name", 'surname' : "last_name"})
#แก้ไขชื่อคอลัมน์โดยใช้ Dictionaries
df_edit_name
```

# Selecting Data

```python
df_lookup = df_edit_name.copy()
#เลือก where หาคอลัมน์ age มากกว่า 100
df_lookup1 = df_lookup[df_lookup["age"] > 100]
df_lookup1 = df_lookup[df_lookup["age"] >= 80]
df_lookup1

df_lookup2 = df_lookup[(df_lookup["age"] > 30) & (df_lookup["age"] < 100)]
df_lookup2
#(&=และ) (|=หรือ) 
df_lookup3 = df_lookup[(df_lookup["country"] == "US") | (df_lookup["country"] == "THA")]
df_lookup3
```

# Unique

```python
df_unique = df_lookup["country"].unique()
df_unique = df_lookup["last_name"].unique()
#เป็นทำข้อมูลที่ไม่ซ้ำเปรียบเหมือน District
df_unique
```

# Pivot

```python
df_pivot = pd.pivot_table(df_lookup,
                index="last_name",
                columns="country",
                values="age",
                aggfunc="count")
df_pivot
```

# FillNaN

```python
df_fillna = df_pivot.fillna(0)
print(len(df_lookup))
df_fillna
```

# Export
